# Top 100 movie titles from AlloCiné

```python
import locale
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import date

locale.setlocale(locale.LC_ALL, 'fr_FR')
```

```python
urls = []
for i in range(1, 8):
    if i == 1:
        urls.append("https://www.allocine.fr/series-tv/")
    else:
        urls.append("https://www.allocine.fr/series-tv/?page={}".format(i))
```

```python
def get_titles(url): 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    selection_class='meta-title'
    title_tags=soup.find_all('h2',{'class':selection_class})
    titles=[]
    for tag in title_tags:
        title = tag.find('a').text
        titles.append(title)       
    return titles
```

```python
top_series_titles = []
for url in urls:
    titles = get_titles(url)
    top_series_titles.extend(titles)
series_data = {'title' : top_series_titles}
pd.set_option("display.max_rows", None)
series_df = pd.DataFrame(series_data)
```

```python
today = date.today().strftime("%d_%m_%Y")
series_df.to_csv("top_series_allocine_{}.tsv".format(today), sep = '\t', index = False)
```
