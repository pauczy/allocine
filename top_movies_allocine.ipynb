# Top 100 movie titles from AlloCiné

```python
import locale
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import date

locale.setlocale(locale.LC_ALL, 'fr_FR')
```

```python
urls = []
for i in range(7):
    if i == 1:
        urls.append("https://www.allocine.fr/films/")
    else:
        urls.append("https://www.allocine.fr/films/?page={}".format(i))
```

```python
def get_movie_titles(url): 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    selection_class='meta-title'
    movie_title_tags=soup.find_all('h2',{'class':selection_class})
    movie_titles=[]
    for tag in movie_title_tags:
        title = tag.find('a').text
        movie_titles.append(title)       
    return movie_titles
```

```python
def get_movie_dates(url): 
    response = requests.get("https://www.allocine.fr/films/")
    soup = BeautifulSoup(response.text, "html.parser")
    selection_class='meta-body-item meta-body-info'
    movie_year_tags=soup.find_all('div',{'class':selection_class})
    movie_years=[]
    for tag in movie_year_tags:
        year = tag.find('span').text
        movie_years.append(year)       
    return movie_years
```

```python
top_movies_titles = []
top_movies_dates = []
for url in urls:
    titles = get_movie_titles(url)
    dates = get_movie_dates(url)
    top_movies_titles.extend(titles)
    top_movies_dates.extend(dates)
movie_data = {'title' : top_movies_titles, 'release date' : top_movies_dates}
pd.set_option("display.max_rows", None)
movies_df = pd.DataFrame(movie_data)
movies_df['release date'] = pd.to_datetime(movies_df['release date'], format='%d %B %Y')
```
